```python
### get file names ###
env = 'local'
files = os.listdir('../coverage-effect/database/')
filenames = []

allwords = ['E_no','111']
allnwords = ['__', 'all', 'ZnCu','CoPt3','CuSn','GaNi','Ru','Zn','110','211','u4','u3']

for i in range(len(files)):
    if all((s in files[i]) for s in allwords):
        if all((s not in files[i]) for s in allnwords):
            filenames.append(files[i])
            print(i, files[i])
```

```python
adsatoms = query('CO_RPBE.traj')
i = 0
alldata = [] # edge_index, features
y_atomint = []
y_spaceint = []
y_allint = []
for file in filenames:
    atoms = init_query(file)
    
    client = MongoClient('localhost', 27017)
    db = client.adsE_database
    collection = db['adsE_method2_new']
    
    dic = {'name': file}
    data = list(collection.find(dic))[0]
    if data['ispredictable']=='yes' and data['isvalid']=='yes':
        print(i,data['name'])
        alldata.append(get_edge_index_and_feature_and_filter(atoms, adsatoms))
        y_atomint.append(data['E_residue/suratom'])
        y_spaceint.append(data['E_int_space']/data['surfatomnum'])
        y_allint.append(data['E_residue/suratom'] + data['E_int_space']/data['surfatomnum'])
        i+=1
```

```python
tmp = np.array(alldata)
edge_index = tmp[:,0]
features = tmp[:,1]
surf_filter = tmp[:,2]
y_atomint = np.array(y_atomint)
y_spaceint = np.array(y_spaceint)
y_allint = np.array(y_allint)
```

```python
data = [edge_index,features,surf_filter,y_atomint, y_spaceint,y_allint]
```
